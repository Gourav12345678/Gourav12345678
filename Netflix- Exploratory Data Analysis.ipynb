```python
import zipfile
import os

# Specify the path to the uploaded zip file on your desktop
zip_file_path = os.path.join(os.path.expanduser("~"), "Desktop", "sample_project", "Netflix_EDA", "Netflix","Netflix_Data.zip")

# Specify the destination directory for extraction
extracted_dir_path = os.path.join(os.path.expanduser("~"), "Desktop", "sample_project", "Netflix_EDA", "Netflix")

# Create the destination directory if it doesn't exist
os.makedirs(extracted_dir_path, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_dir_path)

print(f"File '{zip_file_path}' successfully extracted to '{extracted_dir_path}'.")
```

#  The dataset consist of following columns

- show_id: Gives the information about show id.
- type: Gives information about 2 different unique values one is TV Show and another is Movie.
- title: Gives information about the title of Movie or TV Show.
- director: Gives information about the director who directed the Movie or TV Show.
- cast: Gives information about the cast who plays role in Movie or TV Show.
- country: Gives information about the Name of country.
- date_added: Gives information about the tv shows or movie released.
- release_year: Gives information about the year when Movie or TV Show was released.
- rating: Gives information about the Movie or TV Show are in which category (eg like the movies are only for students, or adults, etc).
- duration: Gives information about the duration of Movie or TV Show.
- listed_in: Gives information about the genre of Movie or TV Show.
- description: Gives information about the description of Movie or TV Show.

```python
country_counts = df['country'].value_counts(normalize=True)[:10]

# Create a pie chart using Matplotlib and Seaborn
plt.figure(figsize=(14.5, 14.5))
sns.set_palette("pastel")  # Set color palette

# Plot the pie chart
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=90,
       textprops={'fontsize':11,'fontweight':'bold'})

# Set plot title
plt.title('Netflix Contribution by Country',fontweight='heavy')

# Show the plot
plt.show()
```

- #### As per above chart United States & India contributes most in Netflix

### Analysing Number of Shows/Movies Added Each Year

```python
plt.figure(figsize=(12, 6))
sns.countplot(df, x='year_added', palette='viridis')
plt.title('Number of Shows/Movies Added Each Year')
plt.show()
```

- #### 2019 gave the highest number of shows folllowed by 2020

 ### Shows/Movies added over the Years

```python
plt.figure(figsize=(12, 6))
sns.heatmap(pd.crosstab(df['year_added'], df['type']), cmap='coolwarm', annot=True,fmt='d')
plt.title('Shows/Movies Added Over the Years')
plt.show()
```

- #### On 2019 highest number of movies & 2020 highest number of TV shows added

### Analysing the content rating

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='rating', order=df['rating'].value_counts().index)
plt.title('Distribution of Content Ratings')
plt.xticks(rotation=45, ha='center')
plt.show()
```

```python
# Analysing listed in feature
df.head()
```

```python
# Analysing Listed_in 

Split the 'listed_in' column into individual genres
genres_split = df['listed_in'].str.split(', ').explode()

# Count the occurrences of each genre
genre_counts = genres_split.value_counts()

#  Create a pie chart
plt.figure(figsize=(10, 10))
genre_counts.head(10).plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Genres Distribution on Netflix')
plt.show()
```

- #### Netflix has highest number of Internationala movies followed by  Dramas and then Comedies.

### Bivariate Analysis

```python
top_countries = df['country'].value_counts().head(5).index
temp_df = df[df['country'].isin(top_countries)]

# Created a stacked bar plot
plt.figure(figsize=(12, 8))
sns.countplot(data=temp_df, x='country', hue='type', palette='viridis')
plt.title('Popularity of Movie and TV Show Types in Top 5 Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.legend(title='Type', loc='upper right')
plt.show()
```

- #### In USA and India 'Movie' show are more famous than 'TV Shows',and in United Kingdom both type are equily popular,and In Japan 'TV shows' are more popular than 'Movie' shows.

```python
top_countries = df['country'].value_counts().head(5).index
temp_df = df[df['country'].isin(top_countries)]

# Created a stacked bar plot
plt.figure(figsize=(12, 8))
sns.countplot(data=temp_df, x='country', hue='rating', palette='viridis')
plt.title('Rati in Top 5 Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.legend(title='Type', loc='upper right')
plt.show()
```

- #### For USA most of the shows are TV-MA rating and for India most of the shows has TV-14 rating
